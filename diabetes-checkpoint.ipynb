```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns ; sns.set()
from matplotlib.pyplot import style ; style.use('ggplot')
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
sns.countplot(df.Outcome)
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x=df.Age,y=df.BMI,hue=df.Outcome)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x=df.Glucose,y=df.BloodPressure,hue=df.Outcome)
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),annot=True,cmap='RdYlGn')
plt.show()
```

```python
sns.scatterplot(x=df.Glucose,y=df.Outcome,hue=df.Outcome)
```

```python
from sklearn.decomposition import PCA
pca = PCA(2)
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
from sklearn.model_selection import train_test_split as tts
x_train , x_test , y_train , y_test = tts(x,y)
pca.fit(x,y)
```

```python
x_pca = pca.transform(x)
```

```python
plt.figure(figsize = (12,10))
sns.scatterplot(x_pca[:,0],x_pca[:,1],hue=df.Outcome)
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(2)
x_tsne = tsne.fit_transform(x,y)
```

```python
plt.figure(figsize = (12,10))
sns.scatterplot(x_tsne[:,0],x_tsne[:,1],hue=df.Outcome)
plt.show()
```

```python
from sklearn.svm import SVC
svm = SVC(gamma=0.01)
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
rf = RFC()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier as GFC
gf = GFC()
gf.fit(x_train,y_train)
gf.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB as MNB
mnb = MNB()
mnb.fit(x_train,y_train)
mnb.score(x_test,y_test)
```

```python
from sklearn.neural_network import MLPClassifier as ML
ml = ML()
ml.fit(x_train,y_train)
ml.score(x_test,y_test)
```

```python
from xgboost import XGBClassifier as XG
xg = XG()
xg.fit(x_train,y_train)
xg.score(x_test,y_test)
```
